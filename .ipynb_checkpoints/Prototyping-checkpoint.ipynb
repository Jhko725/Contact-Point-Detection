```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
data_path = './Data/191109Approach 6.txt'
```

```python
def _GetHeader(header_line):
    index = []
    header = []
    
    for i, word in enumerate(header_line.split('\t')):
        if not word.isspace():
            index.append(i)
            header.append(word.strip())
    return index, header
```

```python
with open(data_path, 'r') as file:
    index, header = _GetHeader(file.readline())
    
    d = []
    amp = []
    phas = []
    z_in = []
    z_out = []
    
    for line in file:
        str_list = line.split('\t')
        d.append(str_list[0])
        amp.append(str_list[1])
        phas.append(str_list[2])
        z_in.append(str_list[index[-2]:index[-1]])
        z_out.append(str_list[index[-1]:-1]) #-1 included to exclude \n
d = np.array(d, dtype = float)
amp = np.array(amp, dtype = float)
phas = np.array(phas, dtype = float)
z_in = np.array(z_in, dtype = float)
z_out = np.array(z_out, dtype = float)
```

```python
fig, ax = plt.subplots(1, 1, figsize = (7, 5))
ax.plot(d, amp, '.k')
ax.grid(ls = '--')
```

```python
z = np.stack([z_in, z_out], axis = -1)
z = z - np.mean(z, axis = 0)
z_fft = np.abs(np.fft.rfft(z, axis = 1))
```

```python
def AveragedSpectra(x, N, axis = -1):
    """
    x : data array of interest
    N : number of samples for averaging
    axis : axis along which to perform the fft
    """
    x_ = np.swapaxes(x, 0, axis)
    x_list = []
    x_len = x_.shape[0]//N
    
    for i in range(N):
        x_list.append(x_[i*x_len:(i+1)*x_len])
    
    xfft_list = [np.abs(np.fft.rfft(x_piece, axis = 0)) for x_piece in x_list]
    xfft = np.mean(np.stack(xfft_list, axis = -1), axis = -1)
    
    return np.swapaxes(xfft, 0, axis)
```

```python
z_fft = AveragedSpectra(z, 50, axis = 1)
```

```python
ind = 700

fig, axes = plt.subplots(1, 2, figsize = (15, 5))
axes[0].plot(z[ind, :, 0], '.k', alpha = 0.3, label = 'Input Signal')
axes[0].plot(z[ind, :, 1], '.r', alpha = 0.3, label = 'Output Signal')
axes[0].grid(ls = '--')
axes[0].legend(loc = 'best')

axes[1].plot(z_fft[ind, :, 0], '.k', alpha = 0.3, label = 'Input Spectra')
axes[1].plot(z_fft[ind, :, 1], '.r', alpha = 0.3, label = 'Output Spectra')
axes[1].grid(ls = '--')
axes[1].legend(loc = 'best')
axes[1].set_yscale('log')
```

```python
fig, ax = plt.subplots(1, 1, figsize = (7, 5))
ax.plot(z_fft[0, :, 1], '.k', alpha = 0.7, label = 'Output Spectra at ind = 0')
ax.plot(z_fft[750, :, 1], '.r', alpha = 0.7, label = 'Output Spectra at ind = 700')
ax.grid(ls = '--')
ax.legend(loc = 'best')
#ax.set_yscale('log')
plt.axhline(y = np.mean(z_fft[0, :, 1]), color = 'k', alpha = 0.5)
plt.axhline(y = np.mean(z_fft[750, :, 1]), color = 'r', alpha = 0.5)

plt.axvline(x = np.argmax(z_fft[0, :, 0]), color = 'b', alpha = 0.5)
```

```python
from sklearn.decomposition import FastICA
ica = FastICA(n_components = 2)

result = ica.fit_transform(z[ind,:,:])
result_fft = np.abs(np.fft.rfft(result, axis = 0))
```

```python
result_fft = AveragedSpectra(result, 50, 0)
```

```python
import numpy as np
from scipy.special import kv
from numba import vectorize, float64

k = 1000 #N/m
Om = 1
Q = 12000
w0 = 32768
a = 1e-3
mu = 8.94e-4 # Dynamic viscosity of water: 8.94mPa*s
rho = 994 # Density of water : 994kg/m^3
```

```python
R*kv(1, R)/kv(0, R)
```

```python
L = 0.0
```

```python
1/((1-Om**2 + 1.0j*Om/Q + 2.0j*np.pi*mu*Om*w*R*kv(1, R)/kv(0, R)/k)*Q)
```

```python
qtf_params = {'a': 1e-3, 'w0': 32768, 'Om': 1, 'Q': 12000, 'k': 1000}

def AmpPhas(L, mu, rho, a, w0, Om, Q, k):
    w = Om*w0
    nu = mu/rho
    R = a*np.sqrt(1.0j*w/nu)
    bessel_part = R*kv(1, R)/kv(0, R)
    result = 1/((1-Om**2 + 1.0j*Om/Q + 2.0j*np.pi*mu*Om*w*L*bessel_part/k)*Q)
    
    return np.abs(result), np.angle(result, deg = True)
```

```python
L = np.linspace(0.0, 1e-4, 1000)
A, theta = AmpPhas(L, mu, rho, **qtf_params)
```

```python
import matplotlib.pyplot as plt
fs = 14
fig, ax1 = plt.subplots(1, 1, figsize = (7, 5))
ax2 = ax1.twinx()
line1 = ax1.plot(L, A, 'k', label = 'Normalized Amplitude')
line2 = ax2.plot(L, theta, 'r', label = 'Phase')
lines = line1+line2
lbs = [l.get_label() for l in lines]
ax1.grid(ls = '--')
ax1.legend(lines, lbs, loc = 'best', fontsize = fs-2)
ax1.set_xlabel('Dipping Depth $L$ (m)', fontsize = fs)
ax1.set_ylabel('Normaliazed Amplitude $A/A_0$', fontsize = fs)
ax2.set_ylabel('Phase $\\theta$ (deg)', color = 'r', fontsize = fs)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
```
